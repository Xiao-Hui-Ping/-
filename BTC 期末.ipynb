```python

#黃金與死亡交叉 買賣訊號
for i in range(1,BTC_input.shape[0]):
    if(BTC_input["K"].iloc[i] > BTC_input["D"].iloc[i]) and (BTC_input["cum_day"].iloc[i-1]==0):
        BTC_input.iloc[i, -3]  = 1
        BTC_input.iloc[i, -2]  = 1
        BTC_input.iloc[i, -1]  = 0
    elif(BTC_input["K"].iloc[i] < BTC_input["D"].iloc[i])and (BTC_input["cum_day"].iloc[i-1]==1):
        BTC_input.iloc[i, -3]  = 0
        BTC_input.iloc[i, -2]  = 0
        BTC_input.iloc[i, -1]  = 1
    elif(BTC_input["cum_day"].iloc[i-1]>0):
        BTC_input.iloc[i, -3]  = 0
        BTC_input.iloc[i, -2]  = 1
        BTC_input.iloc[i, -1]  = 0
```

```python
initial = 10000
for i in range(1, BTC_input.shape[0]):
#     買的訊號
    if(BTC_input["K"].iloc[i] > BTC_input["D"].iloc[i]) and (BTC_input["cum_day"].iloc[i-1]==0):
#         持有數量 = 手中的資本/ 該標的收盤價
        BTC_input.iloc[i, -3] = initial/ BTC_input["Adj Close"].iloc[i]
#     市值 = 持有數量* 該標的收盤價
        BTC_input.iloc[i, -2] = BTC_input["number"].iloc[i]*BTC_input["Adj Close"].iloc[i]
#     期間報酬 = (市值-投入成本)/市值
        BTC_input.iloc[i, -1] = round((BTC_input["value"].iloc[i]-initial)/initial,4)
#         賣的訊號
    elif(BTC_input["K"].iloc[i] < BTC_input["D"].iloc[i])and (BTC_input["cum_day"].iloc[i-1]==1):
#         持有數量為 0 
        BTC_input.iloc[i, -3] = 0
#     市值 = 持有數量* 該標的收盤價
        BTC_input.iloc[i, -2] = BTC_input["number"].iloc[i-1]*BTC_input["Adj Close"].iloc[i]
#     期間報酬 = (市值-投入成本)/市值
        BTC_input.iloc[i, -1] = round((BTC_input["value"].iloc[i]-initial)/initial,4)
#         持有的訊號
    elif(BTC_input["cum_day"].iloc[i-1]>0):
#         持有數量與上一期相同
        BTC_input.iloc[i, -3] = BTC_input.iloc[i-1, -3]
#     市值 = 持有數量* 該標的收盤價
        BTC_input.iloc[i, -2] = BTC_input["number"].iloc[i]*BTC_input["Adj Close"].iloc[i]
#     期間報酬 = (市值-投入成本)/市值
        BTC_input.iloc[i, -1] = round((BTC_input["value"].iloc[i]-initial)/initial,4)
```
